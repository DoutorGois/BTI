```python
import os
import json

data_folder = '/home/jovyan/notebook/BTI/Deep/data'
raw_folder  = os.path.join(data_folder,'raw')
tok_folder  = os.path.join(data_folder,'tokenized')

raw_files   = [os.path.join(raw_folder,fname) for fname in os.listdir(raw_folder)]
filenames   = os.listdir(raw_folder)

text_data   = []
pdf_dict    = dict()

for f,fname in enumerate(filenames):    
    raw_fpath = os.path.join(raw_folder,fname)
    tok_fpath = os.path.join(tok_folder,fname)
    tok_fpath = tok_fpath[:-5] + '.txt'
    
    with open(raw_fpath,'r') as jfile:
        json_data = json.load(jfile)
        if 'AB' in json_data:
            with open(tok_fpath,'w') as tok_file:
                tok_file.write(json_data['AB'])
        else:
            os.remove(raw_fpath)
            
        if 'scholar_pdf' in json_data:
            pdf_dict[json_data['PMID']] = json_data['scholar_pdf']
```

```python
for key in pdf_dict:
    if pdf_dict[key]:
        print(pdf_dict[key][0])
        print('')
```
