<b>História das redes neurais</b><br>


1. Redes neurais artificiais multicamadas já existem há mais de 25 anos, o que há de novo em <i>deep learning</i>?

Desde que surgiu um o modelo perceptrons de múltiplas camadas (Multi-layer perceptrons, MLP) existem bons algoritmos de treino para redes com 01 e até 02 camadas escondidas.

Tais algoritmos, no entanto, não se mostram eficientes para redes com muitas camadas escondidas.

1. A novidade se resume apenas aos novos algoritmos de treinamento?

Não, os novos modelos buscam uma maior aproximação com os mecanismos cerebrais naturais, através de conceitos advinhos da neurociência.

1. EM que sentido ocorrem essa aproximação com modelos naturais?
Basicamente através da extração dos descritores (features) que passa a ser feita pelos modelos, e o uso de novas funções para processamento e propagação dos estímulos recebidos.

Feature visualization of convolutional net trained ImageNet [Zeiler e Fergus 2013]

Link = https://cs.nyu.edu/~fergus/papers/zeilerECCV2014.pdf

1. A aproximação se limita à extração does descritores?

Não, está também nma forma hoje compreendida de como o cérebro faz o fluxo e o processamento dos estímulos que recebe de diferentes órgãos.

2. O cérebro possui uma arquitetura profunda
2. Os processos cognitivos parecem ser profundos
2. O procesamento é distribuido e, geralmente, esparso.
