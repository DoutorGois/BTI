1. O que representa cada dimensão de X?

O vetor <b><i>x</b></i> pertence ao espaço real com <b><i>k</b></i> dimensões; cada dimensão em <b><i>k</b></i> equivale a uma medida descritiva do vetor <b><i>x</b></i>.

1. O que é representado em uma matriz X de dimensões k x n?

Sendo <b><i>X</b></i> uma matriz no espaço real com <b><i>k</b></i> vezes <b><i>n</b></i> dimensões, cada linha <b><i>k</b></i> representa uma medida descritiva da <b><i>n</b></i>-ésima instância do vetor <b><i>x</b></i>.

1. O que representam as matrizes Cov(X) e Corr(X)?

O vetor descritor <b><i>d</b></i> pertence ao espaço real com <b><i>n</b></i> dimensões. Cada dimensão representa a medição do descritor no <b><i>n</b></i>-ésimo vetor <b><i>x</b></i>.

A covariância é um operador binário entre dois vetores do mesmo espaço ($R^n$), calculado como a média do produto dos afastamentos de cada descritor de sua média. Matemáticamente pode ser escrita como: 

\begin{equation}
Cov(x,y) = \sum\limits_{i=1}^{n}  \frac{(x_i-\bar x)(y_i-\bar y)}{n} \\
\text{where } \bar \alpha = \sum\limits_{i=1}^{n}\frac{\alpha}{n}.
\end{equation}

A correlação é uma ferramenta simples, mas poderosa, desenvolvida por Karl Pearson. O $\rho$ de Pearson como também é conhecido pode ser lido e interpretado de várias formas[1]. Dentro do contexto da estatística, $\rho$ é definido como um operador binário de entre dois elementos do mesmo espaço (<b><i>$R^n$</b></i>) calculado como a razão entre a covariância destes dois elementos e o produto de seus desvios padrões.

\begin{equation}
Corr(x,y) =   \frac{Cov(x,y)}{Std(x)*Std(y)}\\
\text{where }Std(\alpha) = \sum\limits_{i=1}^{n}\sqrt[2]{(\alpha_i-\bar \alpha)^2}
\end{equation}

Contudo, a interpretação que mais me agrada é a interpretação geométrica: A correlação de Pearson é o cosseno do angulo formado entre os dois vetores <b><i>x</b></i> e <b><i>y</b></i> no espaço <b><i>$R^n$.</b></i>
<img src="./Angle.png"/>

Seja <b>X</b> $\in R^{n,k}$, cada coluna de <b>X</b> é denotado por $v_{i}$  $\in R^{n}\text{ para }i \in N \text{ e } i  \leq k$. As matrizes de covariância e de correlação são matrizes quadradas no $R^{k,k}$, as quais cada elemento $a_{i,j}\text{ para }i \in N \text{ sendo } i  \leq k\text{ e }j \in N \text{ sendo } j  \leq k$ assume, respectivamente, valores $a_{i,j}=Cov(v_{i},v_{j})$ e $a_{i,j}=Corr(v_{i},v_{j})$.

4. Dado duas variáveis distintas, o que significa dizer que são correlacionadas, descorrelacionadas e independentes?

Utilizando-me da interpretação geométrica, duas variáveis são ditas correlacionadas quando o cosseno do ângulo de seus vetores possuírem valor positivo, descorrelacionadas se o cosseno for negativo e independente se o cosseno for igual a zero. Embora essas definições estejam corretas, a relevância dessa métrica depende tanto de sua magnitude que pode ser despresível, e do valor da probabilidade de refutação da hipótese nula que pode ser não significativa.

5. Explique o que são e para que servem as seguintes ferramentas:<br>

a. PCA<br>

A sigla inglesa PCA extende para Principal Component Analysis - do português Análise de Componentes Principais, ou ACP. É uma mudança de base realizada em elementos do $R^{n,k}$ através da projeção das observações no conjunto de  autovetores da matriz de covariância $R^{k,k}$.

b. ICA<br>

A sigla inglesa ICA extende para Independent Component Analysis - do português Análise de Componentes Independente, ou ACI. É uma mudança de base realizada em cada observação $R^{1,k}$ a matriz $X \in R^{n,k}$ através da projeção das observações no conjunto de  autovetores ($\in R^{k,1}$) da matriz de $COV(X)$.

c. LDA<br>

A sigla inglesa LDA extende para Linear Discriminant Analysis - do português Análise de Discriminante Linear, ou ADL. É uma mudança de base realizada em cada observação $R^{1,k}$ a matriz $X \in R^{n,k}$ através da projeção das observações no conjunto de  autovetores ($\in R^{k,1}$) da matriz de $COV(X)$.

```python
import pandas as pd

Database = pd.read_excel('./DL03_Teste01_Dados.xlsx')
Database = Database.iloc[1:]

Database['Código'] = Database['Código'].astype(int)

Database = Database.set_index('Código')

Covariance = Database.cov()
Correlation = Database.corr()

def plot_frame(data=Database.cov(),title='Covariance'):
    from matplotlib import pyplot as plt
    from matplotlib import cm as colormap

    try:
        ylabel = [str(k)+' - '+key for k,key in enumerate(Covariance.keys())]
    except:
        ylabel = [str(k) for k in range(len(Covariance))]
    ytick  = [x for x in range(0,len(data),1)]

    figure = plt.figure(figsize=(8, 6), dpi=80, facecolor='w', edgecolor='k')
    axis   = figure.add_subplot(111)
    cmap   = colormap.get_cmap('viridis',5)

    plt_cov = axis.imshow(data)
    axis.grid(True)
    axis.set_xticks(ytick)
    axis.set_yticks(ytick)
    axis.set_yticklabels(ylabel,fontsize=15)
    figure.colorbar(plt_cov)
    plt.axis('equal')
    plt.title(title, fontsize=20)
    plt.show()
```

```python
analysis_dict={'key1':[],'key2':[],'alias':[],'alias2':[],'corr':[],'cov':[]}

for kk1,k1 in enumerate(Correlation.keys()):
    for kk2,k2 in enumerate(Correlation.keys()):
        if kk1 < kk2: 
            analysis_dict['key1'].append(k1)
            analysis_dict['key2'].append(k2)
            analysis_dict['alias'].append(k1[:3]+' x '+k2[:3])
            analysis_dict['alias2'].append(str(kk1)+' x '+str(kk2))
            analysis_dict['corr'].append(Correlation[k1][k2])
            analysis_dict['cov'].append(Covariance[k1][k2])
```

```python
from bokeh.plotting import figure, output_file, show
from bokeh.io import output_notebook

output_notebook()

from bokeh.plotting import figure, show, output_file
from bokeh.models import ColumnDataSource, Range1d, LabelSet, Label

p = figure(title='Correlation vs. Covariation')
source = ColumnDataSource(data=analysis_dict)
p.scatter(x='corr', y='cov', size=8, source=source)
labels = LabelSet(x='corr', y='cov',text='alias', x_offset=5, y_offset=5, source=source)
p.add_layout(labels)
p.xaxis[0].axis_label = 'Correlation'
p.yaxis[0].axis_label = 'Covariation'
show(p)
```
