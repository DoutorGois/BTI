```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats.mstats import zscore
%matplotlib inline

n     = 1000

x     = np.random.randn(n)
alpha = 4
beta  = 5
fnc   = lambda a,b: a*x+b
erro  = lambda m,s: m+np.random.randn(n)*s


y     = fnc(alpha,beta)+erro(0,2)
s2    = lambda a,b:(np.sum(y-(a+b*x))**2)/len(y)
s0    = lambda n,w,h: h+w*np.array([x for x in range(n)])/n

n2    = 100*2
a_s   = np.array(s0(n2,4,3))
b_s   = np.array(s0(n2,5,6))

sm    = np.empty([n2,n2])

for i,a in enumerate(a_s):
    for j,b in enumerate(b_s):
        sm[i,j] = s2(a,b)
```

```python
plt.figure(figsize=(10,10))
plt.imshow(zscore(sm,axis=1))
plt.ylim([99,102])
```
