## Axiomas de Incidencia

### Axioma #1

Para uma reta r em um plano existem pontos que pertencem a reta e pontos que n√£o pertencem a reta

```python
from matplotlib.figure import Figure
import matplotlib.pyplot as plt
import numpy as np

plt.figure(num=1, figsize=(13, 8), dpi=80, facecolor='w', edgecolor='k')

X = np.random.rand(1000,2);
reta = lambda x: .25+.5*x
x    = [0,1]
y    = [x for x in map(reta,x)]
x2   = np.random.rand(25,1)
y2   = [x for x in map(reta,x2)]
plt.plot(x,y,'b',linewidth=5)
plt.plot(x2,y2,'or',markeredgecolor='k')
plt.ylim((0,1))
plt.xlim((0,1))
```

### Axioma #2

Para um conjunto de que cont√©m quaisquer dois pontos pertencentes ao plano h√° apenas uma reta que o cont√©m.

```python

A  = [0.2,0.8]
B  = [0.8,0.6]

dx = B[0]-A[0]
dy = B[1]-A[1]

a  = dy/dx
b  = A[1]-A[0]*a
y  = lambda x: a*x+b;

plt.figure(figsize=(13,5))

plt.plot(A[0],A[1],'o',markerfacecolor='b',markersize=20,markeredgecolor='k')
plt.plot(B[0],B[1],'o',markerfacecolor='r',markersize=20,markeredgecolor='k')
plt.plot([0,1],[y(0),y(1)],'k')

plt.xlim((0,1))
plt.ylim((0,1))
```

## Axiomas de Ordem

### Axioma #1

Dados tr√™s pontos pertencentes a uma reta, apenas um est√° entre os outros dois

```python
msize=20
A  = [0.2,0.4]
B  = [0.8,0.4]
C  = [0.5,0.4]

dx = B[0]-A[0]
dy = B[1]-A[1]

a  = dy/dx
b  = A[1]-A[0]*a
y  = lambda x: a*x+b;

plt.figure(figsize=(13,1))
plt.plot([0,1],[y(0),y(1)],'k')

plt.plot(A[0], A[1], 'o', markerfacecolor='b', markersize=msize, markeredgecolor='k')
plt.plot(B[0], B[1], 'o', markerfacecolor='r', markersize=msize, markeredgecolor='k')
plt.plot(C[0], C[1], 'o', markerfacecolor='g', markersize=msize, markeredgecolor='k')

plt.xlim((0,1))
plt.ylim((0,1))
```

### Axioma #2

Assumindo pontos em uma reta, h√° sempre um ponto C localizado entre A e B e um ponto D no qual B est√° localizando entre A e D.

```python
msize=20
A  = [0.2,0.4]
B  = [0.8,0.4]
C  = [0.5,0.4]
D  = [0.9,0.4]

dx = B[0]-A[0]
dy = B[1]-A[1]

a  = dy/dx
b  = A[1]-A[0]*a
y  = lambda x: a*x+b;

plt.figure(figsize=(13,1))
plt.plot([0,1],[y(0),y(1)],'k')

plt.plot(A[0], A[1], 'o', markerfacecolor='b', markersize=msize, markeredgecolor='k')
plt.plot(B[0], B[1], 'o', markerfacecolor='r', markersize=msize, markeredgecolor='k')
plt.plot(C[0], C[1], 'o', markerfacecolor='g', markersize=msize, markeredgecolor='k')
plt.plot(D[0], D[1], 'o', markerfacecolor='brown', markersize=msize, markeredgecolor='k')

plt.xlim((0,1))
plt.ylim((0,1))
```
