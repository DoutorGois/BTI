```c++
// José Henrique Targino Dias Góis
// 20180057876
#include<iostream>
#include<string>
#import<math.h>

int cannonball(int h);
int power(int n, int k);
int digitSum(int n);
int digitSum2(int n);
int gdc(int x, int y);
int digitalRoot(int n);
int pascal(int n, int k);
bool find(int *a, int *b, int v);
int * unique(int *a, int *b);
bool palindrome(std::string text);

int cannonball(int h){
    if(h==1)
        return 1;
    return h*h + cannonball(h-1);
}

int power(int n, int k){
    if(k==0)
        return 1;
    return n*power(n,k-1);
}

int digitSum(int n){
    if(0==(n/10))
        return n;
    return n-10*(n/10) + digitSum(n/10);
}

int digitSum2(int n){
    int k = floor(log10(n));
    if (k==0)
        return n;
    int digit = n/power(10,k);
    return digit + digitSum2(n-digit*power(10,k));
}

int gdc(int x, int y){
    if(y==0)
        return x;
    return gdc(y,x%y);
}

int digitalRoot(int n){
    if(0==(n/10))
        return n;
    return digitalRoot(digitSum(n));
}

int pascal(int n, int k){
    if(k>n)
        return 0;
    if(n==0 | k==0 | k==n)
        return 1;
    return pascal(n-1,k-1)+pascal(n-1,k);
}

bool find(int *a, int *b, int v){
    if(*a==v)
        return true;
    if(a==b)
        return false;
    return find(a+1,b,v);
}

int * unique(int *a, int *b){
    return b
}

bool palindrome(std::string text){
    if (text.size()==1)
        return true;
    if (((*(text.begin())) ^~ (*(text.end()-1))) == -1)
        return palindrome(text.substr(1,text.size()-2));
    return false;
}
```
